#Null Value Drops

```python
df["country_clean"] = [x.split(",")[0] if pd.notnull(x) else x for x in df.country ]
df["genre_clean"] = [x.split(",")[0] if pd.notnull(x) else x for x in df.genre ]
df.info()
```

```python
Netclean_df = df[df["type"] != "TV Show"]
Netclean_df.head()
```

```python
Netclean_df['duration_clean'] = Netclean_df['duration'].str.split().str[0]
Netclean_df
```
